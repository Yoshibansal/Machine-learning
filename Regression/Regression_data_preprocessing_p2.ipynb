```python
import numpy as np
import pandas as pd
```

```python
path = 'https://raw.githubusercontent.com/Yoshibansal/Machine-learning/main/Regression/communities.data'
```

```python
raw_df = pd.read_csv(path, sep=',', header=None)
pd.set_option('display.max_columns', None)
raw_df.head()
```

```python
raw_df.shape
```

```python
df = raw_df.replace("?", np.nan)
```

```python
pd.set_option('display.max_rows', None)
nul_col = df.isnull().sum()
```

```python
print(nul_col[nul_col > 0])
```

```python
nan_cols = [i for i in df.columns if df[i].isnull().any()]
```

```python
nan_cols
```

```python
print("Total values in 3rd coloumn: ", len(df[3]))
print("Unique values in 3rd coloumn: ", len(df[3].unique()))
```

```python
df.head()
```

```python
df[nan_cols] = df[nan_cols].fillna(method='ffill')
```

```python
df.head()
```

```python
df = df.replace(np.nan, 1);
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()

df[[0,1,2,4]] = pd.DataFrame(scale.fit_transform(df[[0,1,2,4]].values), columns=[0,1,2,4], index=df.index)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()

df[3] = enc.fit_transform(df[3])
```

```python
df.head()
```

```python
df = df.to_numpy()

X = df[:, :-1]
y = df[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
#@title Degree
degree = 2 #@param {type:"slider", min:1, max:4, step:1}

model = make_pipeline(PolynomialFeatures(degree),LinearRegression())
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
r2_train = r2_score(y_train, y_train_pred)
print("RMSE train: ", rmse_train)
print("R2 train: ", r2_train)
```

```python
y_test_pred = model.predict(X_test)
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
r2_test = r2_score(y_test, y_test_pred)
print("RMSE test: ", rmse_test)
print("R2 test: ", r2_test)
```

```python
print("Accuracy/Score (Train): ", model.score(X_train, y_train))
```

Linear Regression

```python
lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

print("MSE: ", mean_squared_error(y_test, y_pred))
```

```python
print("Accuracy/Score (Train): ", lr.score(X_train, y_train))
# print("Accuracy/Score (Test): ", lr.score(X_test, y_test))
```

Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()

regressor.fit(X_train, y_train)
```

```python
print("Accuracy/Score (Train): ", regressor.score(X_train, y_train))
# print("Accuracy/Score (Test): ", regressor.score(X_test, y_test))
```
